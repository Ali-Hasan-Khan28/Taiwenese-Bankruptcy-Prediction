```python
from scipy.stats import t

# read CSV file
df = pd.read_csv('data.csv')

# select two columns
col1 = df[' ROA(C) before interest and depreciation before interest']
col2 = df[' ROA(A) before interest and % after tax']

# calculate the mean and standard deviation of each column
mean1 = np.mean(col1)
mean2 = np.mean(col2)
std1 = np.std(col1)
std2 = np.std(col2)

# calculate the pooled standard deviation
n1 = len(col1)
n2 = len(col2)
pooled_std = np.sqrt(((n1 - 1) * std1**2 + (n2 - 1) * std2**2) / (n1 + n2 - 2))

# calculate the t-statistic
t_stat = (mean1 - mean2) / (pooled_std * np.sqrt(1/n1 + 1/n2))

# calculate the degrees of freedom
df = n1 + n2 - 2

# calculate the p-value
p_value = 2 * (1 - t.cdf(abs(t_stat), df))

# print the resulting t-statistic and p-value
print('t-statistic:', t_stat)
print('p-value:', p_value)
```

it means null  hypothesis is true

A negative t-statistic means that the mean of the first group is lower than the mean of the second group.
